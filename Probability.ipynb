```python
import scipy.stats as stats
import numpy as np
import pylab as pl
```

```python
age = [9,13,18,19,20,21,24,25,27,28,30,30,31,32,33,33,34,34,35,35,36,37,38,39,40,43,44,45,50,51,53,55,58,60,65]
fit = stats.norm.pdf(age, np.mean(age), np.std(age))  #this is a fitting indeed
print(fit)
pl.plot(age,fit,'-or')

pl.hist(age,normed=True)      #use this to draw histogram of your data

pl.show()
```
